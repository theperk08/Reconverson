```python

import pandas as pd
from sklearn.neighbors import NearestNeighbors

df_initial = pd.read_csv('Df_plus_Id.csv')


df_genres = df_initial['genre'].str.get_dummies(sep=',').astype(bool)
df = pd.concat([df_initial, df_genres], axis=1)

# dico des genres similaires
dic = {'Movie' : '', 'R&B': '', 'Alternative': '', 'Country': '', 'Dance': '', 'Electronic': '',
       'Anime': '', 'Folk': '', 'Blues': '', 'Opera': '', 'Hip-Hop': '', "Children's Music": '',
       'Children’s Music': '', 'Rap': '', 'Indie': '', 'Classical': '', 'Pop': '', 'Reggae': '',
       'Reggaeton': '', 'Jazz': '', 'Rock': '', 'Ska': '', 'Comedy': '', 'Soul': '', 'Soundtrack': '',
       'World': ''}

columns = df_initial.columns.values
#titre sélectionné
titre = "Pour que tu m'aimes encore"

# on récupère la(les) ligne(s) du titre
titre_choisi = df_initial[df_initial['track_name'] == titre]
#print(titre_choisi.iloc[:,:3])

# mais on sélectionne que les colonnes sur lesquelles fitter
titre_choisi_fit = titre_choisi.iloc[0:1, 4:]

# on récupère le genre
genre_choisi = titre_choisi.iloc[0,0]
#print('\nGenre choisi :', genre_choisi)

# on ne conserve du dataset que les lignes ayant le même genre
df_cut = df_initial[df_initial['genre'] == genre_choisi]
#print(df_cut)

# on sélectionne les colonnes sur lesquelles fitter
X = df_cut[columns[4:]]

# on entraîne le modèle
distanceKNN = NearestNeighbors(n_neighbors= 10).fit(X)
neighbors = distanceKNN.kneighbors(titre_choisi_fit)
# matrice
#print("Matrice des distances et des voisins:\n", neighbors)

# récupération des titres, genres et artistes les plus proches
titres_bons = df_cut.iloc[neighbors[1][0][1:], 2].values
genres_bons =  df_cut.iloc[neighbors[1][0][1:], 0].values
artistes_bons = df_cut.iloc[neighbors[1][0][1:], 1].values

# affichage des résultats
#print()
for  titre_bon, genre_bon, artiste_bon in zip(titres_bons, genres_bons, artistes_bons):
    print('Titre : {0:25}\n  du genre :{1:10}\n    de l\'artiste : {2}\n'.format(titre_bon, genre_bon,artiste_bon))
```
